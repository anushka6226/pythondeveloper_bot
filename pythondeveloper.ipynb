```python
import datetime
import logging

# Setup logging to a file (Assignment Requirement)
logging.basicConfig(filename='bot.log', level=logging.INFO, format='%(asctime)s - %(message)s')

class MockBinanceClient:
    """A fake Binance client that simulates trading without API keys."""

    def __init__(self):
        self.balance = 15000.0  # Start with fake 15,000 USDT
        print("Mock Client Initialized: No API Keys required.")

    def futures_account_balance(self):
        # Returns a fake balance structure that looks like the real API
        return [{"asset": "USDT", "balance": str(self.balance)}]

    def futures_create_order(self, **params):
        # Simulates placing an order and logs it
        symbol = params.get('symbol')
        side = params.get('side')
        qty = params.get('quantity')
        price = params.get('price', "Market")

        log_msg = f"MOCK TRADE: {side} {qty} {symbol} at {price}"
        print(f"{log_msg}")
        logging.info(log_msg) # This writes to your bot.log file

        return {"status": "NEW", "orderId": 12345, "clientOrderId": "mock_id"}

# Initialize the mock instead of the real one
client = MockBinanceClient()
```

```python
def my_trading_bot(symbol, side, quantity):
    print(f"Bot is analyzing {symbol}...")
    # This calls our fake client instead of the real Binance
    response = client.futures_create_order(symbol=symbol, side=side, quantity=quantity)
    return response

# Execute a fake trade
my_trading_bot("BTCUSDT", "BUY", 0.001)
```

```python
# 1. Define a function that places a trade and prints the result
def run_trade_test(symbol, side, order_type, quantity, price=None):
    print(f"--- Processing {side} Order for {symbol} ---")

    # This calls the Mock Client we built in the previous step
    response = client.futures_create_order(
        symbol=symbol,
        side=side,
        type=order_type,
        quantity=quantity,
        price=price
    )

    return response

# 2. Run a MARKET BUY trade
run_trade_test("BTCUSDT", "BUY", "MARKET", 0.005)

# 3. Run a LIMIT SELL trade
run_trade_test("ETHUSDT", "SELL", "LIMIT", 0.1, price=3500)
```
